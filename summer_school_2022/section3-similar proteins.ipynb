### PDBe Similar proteins for a given protein

This tutorial will guide you through searching PDBe programmatically.


First we will import the code which will do the work
Run the cell below - by pressing the green play button.

```python
import sys
sys.path.insert(0,'..')
from python_modules.api_modules import explode_dataset, pandas_dataset, get_similar_protein_data
import pandas as pd
```

Now we will find all the proteins with 85 % or more sequence identity to SARS-CoV-2 Spike protein (Uniprot accession: P0DTC2).

```python
uniprot_accession = 'P0DTC2'
uniprot_accession
identity_cutoff = 85
results = get_similar_protein_data(uniprot_accession,identity_cutoff) 
df2 = explode_dataset(result=results[uniprot_accession], column_to_explode='mapped_segment')
data = pd.json_normalize(df2['mapped_segment'])
df = df2.join(data).drop(columns='mapped_segment')
df.head()
```

We can save the results to a CSV file which we can load into excel

```python
df.to_csv("similar_proteins_project_aims_8.csv")
```

We can also filter this dataframe of similar proteins based on various column values like sequence_identity, taxid, species. For example, we can filter all the similar proteins from SARS (taxid= 694009) as shown below. 

```python
filter_df = df.loc[df['taxid'] == 694009] 
filter_df.head()
```
