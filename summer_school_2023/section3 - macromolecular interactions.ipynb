### PDBe Macromolecular Interactions for a given protein

This tutorial will guide you through searching PDBe programmatically for finding macromolecular interactions.


First we will import the code which will do the work
Run the cell below - by pressing the green play button.

```python
import pandas as pd
from pprint import pprint
import sys
sys.path.insert(0,'..')
from python_modules.api_modules import explode_dataset, get_macromolecule_interaction_data
```

```python

uniprot_accession = "P22303"
```

Get macromolecules which interact with the UniProt

```python
interaction_data = get_macromolecule_interaction_data(uniprot_accession=uniprot_accession)
```

```python
pprint(interaction_data[0])
```

```python
df2 = explode_dataset(result=interaction_data, column_to_explode='interactingPDBEntries')
```

```python
df2
```

Get all the unique macromolecules interacting with our protein

```python
df2['interaction_accession'].unique()
```

Some post processing is required to separating interactingPDBEntries into separate columns. Here we convert interactingPDBEntries semi-structured JSON data format in df2 column into a flat table.

```python
data = pd.json_normalize(df2['interactingPDBEntries'])
df3 = df2.join(data).drop(columns='interactingPDBEntries')
```

```python
df3
```

startIndex and endIndex are the UniProt residue number, so we'll make a new column called residue_number and copy the startIndex there. We are also going to "count" the number of results - so we'll make a dummy count column to store it in.

```python
df3['residue_number'] = df3['startIndex']
df3['count'] = df3['pdbId']
```

Now we are ready to use the data.
